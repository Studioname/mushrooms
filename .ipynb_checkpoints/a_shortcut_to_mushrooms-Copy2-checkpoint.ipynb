## 1. Encode the Data

Let's encode the data. This will convert the datatypes in our datasets into the kinds a machine learning model can use. We'll use a label encoder for this.

```python
label_encoder = LabelEncoder()
```

```python
for col in df:
    df[col] = label_encoder.fit_transform(df[col])
```

```python
df
```

#### Note on Class label

0 is edible and 1 is poisonous.

## 1. Check for Null Values

```python
df.isnull().sum()
```

```python
# display transpose of dataframe details

df.describe().T
```

Veil-types looks a bit suspicious. Let's see how many unique values there are in that column.

```python
print(df['veil-type'].value_counts())
```

As we can see, there is only one unique value in veil-type. That makes it a constant, which can interfere with a models ability to predict regression vectors. We'll drop it here because there's nothing we can do with it. We can't split it because it's a single value, or remodel it in some way, so we'll drop it.

```python
df = df.drop(['veil-type'], axis=1)
```

```python
df.head()
```

## 2. Draft Model

```python
y = df['class']
x = df.drop(['class'], axis=1)
x.head()
```

```python
x.shape, y.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=100)
```

```python
x_train.shape, y_train.shape
```

```python
svc = SVC().fit(x_train, y_train)
svc.score(x_train, y_train), svc.score(x_test, y_test)
```

```python
k = 10
kf = KFold(n_splits=k, random_state=None)
```

```python
# random forest classifier with k fold cross val
rfc = RandomForestClassifier(n_estimators=50, max_features=5, max_depth=7, random_state=100).fit(x_train, y_train)
rfc.score(x_train, y_train), rfc.score(x_test, y_test)
```

```python
print(cross_val_score(rfc, x, y, cv = kf))
```

Wow! This dataset was a pleasure to work with, but we didn't have to do much to get a great model.

```python
y_pred = svc.predict(x_test)
```

```python
conf = confusion_matrix(y_test, y_pred)
conf
```

```python
print(classification_report(y_test, y_pred))
```

```python
conf_dsp = ConfusionMatrixDisplay(conf, display_labels = [True, False])
conf_dsp.plot()
```
